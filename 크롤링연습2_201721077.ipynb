### 1.2. Text Crawling with for loop
css selector의 규칙을 찾자!

* 1번째 제목: #tab-1 > div > dl > dd:nth-child(1) > a
* 2번째 제목: #tab-1 > div > dl > dd:nth-child(2) > a
<br>. . .

* 5번째 제목: #tab-1 > div > dl > dd:nth-child(5) > a

```python
# chrome driver 설정
driver = webdriver.Chrome('C:/Users/LG/Downloads/chromedriver_win32/chromedriver.exe')
driver.implicitly_wait(10)

url = "https://www.bikeseoul.com/customer/opinionBoard/opinionBoardList.do?currentPageNo=1"

driver.get(url)
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

# title crawling
# css selector의 규칙과 for문을 사용하여 제목 목록을 출력하였다.
for i in range(1, 6):
    title = WebDriverWait(driver, 5) \
        .until(EC.presence_of_element_located((By.CSS_SELECTOR, "#tab-1 > div > dl > dd:nth-child(" + str(i) + ") > a")))
    print("Title: {}".format(title.text))
```

### 1.4. Text Crawling including contents
* 1.2처럼 for문과 함께 써보자! (첫 페이지 5개의 글에 대해 title과 content를 크롤링

* 1.3에서 사용한 click & back을 활용하자

<span style="color:blue"> * 따릉이 시민의견 게시판 사이트의 문의사항 페이지에 변동이 생긴 것 같아 임의로 공지사항 제목과 내용을 크롤링 </span>

```python
 # chrome driver 설정
driver = webdriver.Chrome('C:/Users/LG/Downloads/chromedriver_win32/chromedriver.exe')
driver.implicitly_wait(10)

url = "https://www.bikeseoul.com/customer/notice/noticeList.do?currentPageNo=1" # 공지사항 url
driver.get(url)
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

# 빈 리스트 변수
title_list = []
content_list = []

# title crawling
for i in range(2, 11):
    # click
    click_element = WebDriverWait(driver, 5) \
        .until(EC.presence_of_element_located((By.CSS_SELECTOR, "#sub > div.container > div > div > div.board_box > table > tbody > tr:nth-child("+ str(i) +") > td.left > a")))
    click_element.click()    

# title crawling : 제목 추출
    title = WebDriverWait(driver, 5).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, "#sub > div.container > div > div > div.read_box > dl > dt > p")))
    title_list.append(title.text)
  
    
    # content crawling : 내용 추출
    content = WebDriverWait(driver, 5).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, "#sub > div.container > div > div > div.read_box > dl > dd")))
    content_list.append(content.text)

    # back
    driver.back()

print(title_list)
print(content_list)
```

### 1.5. Text Crawling with double for loop
1Page URL: https://www.bikeseoul.com/customer/opinionBoard/opinionBoardList.do?currentPageNo=1

* 다음 page의 URL은?
<br>2Page URL: https://www.bikeseoul.com/customer/opinionBoard/opinionBoardList.do?currentPageNo=2

PageNo를 for loop 변수로 넣어주면, page도 변경하면서 모든 페이지에 대한 title과 content를 크롤링할 수 있지 않을까?

```python
 # chrome driver 설정
driver = webdriver.Chrome('C:/Users/LG/Downloads/chromedriver_win32/chromedriver.exe')
driver.implicitly_wait(10)

# 빈 리스트 변수
title_list = []
content_list = []

# 1page ~ 5page 크롤링 : for문을 이용해서 1페이지부터 5페이지에 있는 제목과 내용을 크롤링한다.
for page in range(1, 6):
    
    url = "https://www.bikeseoul.com/customer/notice/noticeList.do?currentPageNo="+ str(page) 
    driver.get(url)
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')



# title crawling
    for i in range(2, 11):
        # click
        click_element = WebDriverWait(driver, 5) \
            .until(EC.presence_of_element_located((By.CSS_SELECTOR, "#sub > div.container > div > div > div.board_box > table > tbody > tr:nth-child("+ str(i) +") > td.left > a")))
        click_element.click()    
    
    # title crawling
        title = WebDriverWait(driver, 5).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "#sub > div.container > div > div > div.read_box > dl > dt > p")))
        title_list.append(title.text)


        # content crawling
        content = WebDriverWait(driver, 5).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "#sub > div.container > div > div > div.read_box > dl > dd")))
        content_list.append(content.text)

        # back
        driver.back()

    
    
# 결과 변수
raw_result2 = {'title': title_list,
          'content': content_list}

result2 = pd.DataFrame(raw_result2)

# csv 파일로 save
result2.to_csv("BikeTest2", mode='w')

# driver 종료
driver.quit()  
```

### 1.6. Text Crawling with function (심화)

```python
def BikeDataCrawler(starting_page, end_page):
    # chrome driver 설정
    driver = webdriver.Chrome('C:/Users/LG/Downloads/chromedriver_win32/chromedriver.exe')
    driver.implicitly_wait(10)

    # 빈 리스트 변수
    title_list = []
    content_list = []

    # url에 page가 있는 특성에 따라 for문 생성
    for page in range(starting_page, end_page+1):

        url = "https://www.bikeseoul.com/customer/notice/noticeList.do?currentPageNo="+ str(page) 
        driver.get(url)
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')



     # 페이지마다 5개 글 크롤링
        for i in range(2, 7):
            # click
            click_element = WebDriverWait(driver, 5) \
                .until(EC.presence_of_element_located((By.CSS_SELECTOR, "#sub > div.container > div > div > div.board_box > table > tbody > tr:nth-child("+ str(i) +") > td.left > a")))
            click_element.click()    
        
        # title crawling
            title = WebDriverWait(driver, 5).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, "#sub > div.container > div > div > div.read_box > dl > dt > p")))
            title_list.append(title.text)


            # content crawling
            content = WebDriverWait(driver, 5).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, "#sub > div.container > div > div > div.read_box > dl > dd")))
            content_list.append(content.text)

            # back
            driver.back()

        # page check  
        print(str(page)+"페이지 크롤링 완료!") #한 페이지의 for문이 끝날 때마다 '크롤링 완료'메세지 띄우기

    # 결과 변수
    raw_result = {'title': title_list,
              'content': content_list}

    result = pd.DataFrame(raw_result)
    
    # csv 파일로 save
    result.to_csv("BikeData.csv", mode='w')
    
    # driver 종료
    driver.quit()
    
    # return
    return result   
```
